```
# file to extract. The path below is only an example must be fixed before using
filepath = "/content/drive/My Drive/Hematoma Expansion Cohorts/Final Images/dataset.tar.gz"

tar = tarfile.open(filepath) 
# content = tar.getmembers()
# print(content)

# change extract_to to the directory path where you want the .tar file to be extract. Default is in the same location as the .tar file
extract_to = os.path.dirname(filepath)
extract_to = os.path.join(extract_to, 'dataset')

if not os.path.exists(extract_to):
  os.makedirs(extract_to)
```

```
result = None
while result is None:
  try:
    tar.extractall(path=extract_to)
    tar.close()
    result = 'done!'
  except:
    print('failed...')
    pass

print(result)
```
